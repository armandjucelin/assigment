# Step 2:
Create the following columns in the data:
1. 24 week trailing high
2. 24 week trailing low
3. 12 week trailing high
4. 12 week trailing low

```python
df['Date1'] = pd.to_datetime(df['Date'])
df['week_start'] = df['Date1'] - pd.to_timedelta((df['Date1'].dt.day - 1) % 7, unit='d')
df['week_finish'] = df['week_start'] + pd.Timedelta('6D')
df['week'] = df['week_start'].dt.strftime('%m/%d/%Y').astype(str) + '-' + df['week_finish'].dt.strftime('%m/%d/%Y').astype(str)
data_1 = df.groupby([ 'week_start', 'week'])['High'].sum().reset_index([0,1]).reset_index(drop=True)
data_1 = data_1.loc[::-1]
data_1 =data_1[:23]
data_1
```

```python
#24 week trailing high
week_trailing_high_24 = data_1['High'].max()
week_trailing_high_24
```

```python
data_2 = df.groupby([ 'week_start', 'week'])['Low'].sum().reset_index([0,1]).reset_index(drop=True)
data_2 = data_2.loc[::-1]
data_2 = data_2[:23]
data_2
```

```python
#24 week trailing low
week_trailing_low_24 = data_2['Low'].max()
week_trailing_low_24
```
